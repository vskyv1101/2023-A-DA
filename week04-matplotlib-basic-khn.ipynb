## 마커, 선 너비

```python
import matplotlib.pyplot as plt

plt.plot(range(1, 6, 1), range(50, 9, -10), color='g', ls=':', marker = '^', lw = '2', label='birth rate') # x, y
plt.plot(range(1, 6, 1), range(10, 51, 10), color='skyblue', label='descending')
plt.plot(range(1, 6, 1), [10, 27, 35, 42, 49], color='m', marker = 'd', ls='-.', lw = '3', label='nvidia') # x, y
plt.plot(range(1, 6, 1), [50, 30, 26, 21, 15], color='y', label='AMD')
plt.title('legend')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

N = 8
y = np.zeros(N)

x1 = np.linspace(0, 10, N, endpoint=True) # 전체 8개, 7등분
x2 = np.linspace(0, 10, N, endpoint=False)# 전체 8개, 8등분
plt.plot(x1, y + 0.5, '-o', label='endpoint=True')
plt.plot(x2, y, '.', label='endpoint=False')

plt.legend()
plt.ylim([-0.5, 1])
plt.show()
```
