```python
import csv
import matplotlib.pyplot as plt

f = open('subwaytime.csv', encoding='cp949')
data = csv.reader(f)
header = next(data)
next(data)

max_people = 0
max_station = ''
t = int(input('몇 시의 승차 인원이 가장 많은 역이 궁금하세요?: '))

for row in data:
    for i in range(4, 52):
        row[i] = row[i].replace(',', "")
    
    row[4:] = map(int, row[4:])
    people = row[2 * t - 4]
    if people > max_people:
        max_people = people
        max_station = row[3] + '(' + row[1] + ')'
    
f.close()

print(f'입력한 시간 : {t}')
print(f'{max_station} : {max_people}명')
```

```python

import csv
import matplotlib.pyplot as plt

f = open('subwaytime.csv', encoding='cp949')
data = csv.reader(f)
header = next(data)
next(data)

s_in = [0] * 24 # 승차 리스트 초기화
s_out = [0] * 24 # 하차 리스트 초기화

for row in data:
    for i in range(4, 52):
        row[i] = row[i].replace(',', "")
    
    row[4:] = map(int, row[4:])
    for i in range(24):
        s_in[i] = s_in[i] + row[2*i+4]
        s_out[i] = s_out[i] + row[2*i+5]

f.close()

plt.rc('font', family='Malgun Gothic')
plt.title('지하철 시간대별 승하차 인원 추이')
plt.plot(s_in, label='승차')
plt.plot(s_out, label='승차')
plt.legend()
plt.grid(True)
plt.xticks(range(24), range(4, 28))
plt.show()
```
