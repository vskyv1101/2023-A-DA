```python
import matplotlib.pyplot as plt
import numpy as np
import csv

f = open('age.csv', encoding='cp949')
data = csv.reader(f)

header = next(data)

name = input('인구 구조가 알고 싶은 지역의 이름(읍면동 단위)을 입력해주세요: ')

for row in data:
    if name in row[0]:
        home = np.array(row[3:], dtype=int)
    if '신도림' in row[0]:
        homeB = np.array(row[3:], dtype=int)

f.close()

plt.rc('font', family='Malgun Gothic')
plt.plot(home, label=name)
plt.plot(homeB, label='신도림')
plt.title(name + '지역의 인구 구조')
plt.grid(True)
plt.xlabel('나이')
plt.ylabel('인구수')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import csv

f = open('age.csv', encoding='cp949')
data = csv.reader(f)

header = next(data)

name = input('인구 구조가 알고 싶은 지역의 이름(읍면동 단위)을 입력해주세요: ')

for row in data:
    for i in range(102):
        row[i+2] = int(row[i+2].replace(',', ""))
    if name in row[0]:
        home = np.array(row[3:], dtype=int) / row[2]
    if '신도림' in row[0]:
        away = np.array(row[3:], dtype=int) / row[2]

f.close()

plt.rc('font', family='Malgun Gothic')
plt.plot(home, label=name)
plt.plot(away, label='신도림')
plt.title(name + ' 지역과 신도림 지역의 인구 구조 비교')
plt.grid(True)
plt.xlabel('나이')
plt.ylabel('인구 비율')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import csv

f = open('age.csv', encoding='cp949')
data = csv.reader(f)
header = next(data)
data = list(data)

min_val = 1
result_name = ''
result = 0

name = input('인구 구조가 알고 싶은 지역의 이름(읍면동 단위)을 입력해주세요: ')
for row in data:
    for i in range(102):
        row[i+2] = int(row[i+2].replace(',', ""))
    if name in row[0]:
        home = np.array(row[3:], dtype=int) / row[2]

for row in data:
    if row[2] == 0:
        continue

    if name not in row[0]:
        away = np.array(row[3:], dtype=int) / row[2]
        s = np.sum((home-away)**2)
        if s < min_val:
            min_val = s
            result_name = row[0]
            result = away

f.close()

plt.plot(home, label=name)
plt.plot(result)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import csv

f = open('age.csv', encoding='cp949')
data = csv.reader(f)
header = next(data)
data = list(data)

min_val = 1
result_name = ''
result = 0

name = input('인구 구조가 알고 싶은 지역의 이름(읍면동 단위)을 입력해주세요: ')
for row in data:
    for i in range(102):
        row[i+2] = int(row[i+2].replace(',', ""))
    if name in row[0]:
        home = np.array(row[3:], dtype=int) / row[2]

for row in data:
    if name not in row[0]:
        if row[2] == 0:
            continue

        away = np.array(row[3:], dtype=int) / row[2]
        s = np.sum((home-away)**2)
        if s < min_val:
            min_val = s
            result_name = row[0]
            result = away

f.close()

plt.rc('font', family='Malgun Gothic')
plt.title(name + ' 지역과 가장 비슷한 인구 구조를 가진 지역: ' + result_name)
plt.plot(home, label=name)
plt.plot(result, label=result_name)
plt.grid(True)
plt.xlabel('나이')
plt.ylabel('인구 비율')
plt.legend()
plt.show()
```
