```python
df = df.replace(',','',regex=True)
df = df.apply(pd.to_numeric)
```

```python
df = df.div(df['2022년08월_계_총인구수'], axis='index')
df.head(3)
```

```python
del df['2022년08월_계_총인구수'], df['2022년08월_계_연령구간인구수']
df.head(3)
```

```python
name=input('원하는 지역의 이름을 입력하세요 : ')
a= df.index.str.contains(name)
df2 = df[a]
x = df.sub(df2.iloc[0],axis='columns')
y = np.power(x,2)
z = y.sum(axis='columns')
z=z[z[:] !=0]
i = z.sort_values().index[:5]

plt.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
df.loc[i].T.plot(figsize=(10,6))
plt.xlabel('나이')
plt.ylabel('인구 비율')
plt.grid(True)
plt.show()
```
